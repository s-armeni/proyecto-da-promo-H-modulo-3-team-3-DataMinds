```python
#codigo de MArga para limpiar los nombres de la columnaas
colum_limpias=[]
for columna in df_col_1:
    patron='[A-Z][^A-Z]*'
    columna_limpia=re.findall(patron,columna)
    columna_limpia1='_'.join(columna_limpia)
    columna_min=columna_limpia1.lower()
    colum_limpias.append(columna_min)
df_col_1= colum_limpias
df_col_1
```

```python
#pasamos la primera columna que es un objeto a numero entero
#df_col_1
#df_col_1['age'] = pd.to_numeric(df_col_1['age'])
# no se puede pasar porque no todos los valores son numeros.
# convertir los valores que no son nmumeros a Nan
# Usar pd.to_numeric con errors='coerce' para convertir valores no numéricos a NaN
df_col_1['age'] = pd.to_numeric(df_col_1['age'], errors='coerce').astype('Int64')

# Ahora los valores no numéricos son NaN. Si se desea, estos NaN se pueden rellenar o eliminar
# Rellenar NaN con un valor específico, por ejemplo, la media
df_col_1['age'].fillna(df_col_1['age'].mean(), inplace=True)
```
