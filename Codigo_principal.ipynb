```python
# OTRA MANERA DE HACERLO:

replacement_dict = {'yes': True, 'Yes': True,'no': False,'No': False,'true': True,'false': False, "1": True, "0": False}


df_silvia['remote_work'].replace(replacement_dict)
```

```python
# OTRA MANERA DE HACERLO DENTRO DE UNA FUNCION:

def convierto_boolean(remote_work):

    if remote_work in ('True','yes', 'Yes', '1', 'true'):

        return 'Yes' 
    
    elif remote_work in ('False','no', 'No', '0', 'false'):
        return 'No'
    
    else:
        return np.nan
```

```python
df_silvia["remote_work"]=df_silvia["remote_work"].apply(convierto_boolean)
```
